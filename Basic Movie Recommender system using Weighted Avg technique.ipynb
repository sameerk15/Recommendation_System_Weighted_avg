```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
credits_df=pd.read_csv('tmdb_5000_credits.csv')
credits_df.head()
```

```python
movies_df=pd.read_csv('tmdb_5000_movies.csv')  #vote_average will be the key
movies_df.head()
```

```python
movies_df.shape
```

```python
credits_df.shape
```

```python
#we will combine both the datasets based on 'id' feature as id is present in both datasets
```

```python
credit_column_renamed=credits_df.rename(index=str, columns={'movie_id':'id'})
```

```python
movies_df_merge=movies_df.merge(credit_column_renamed, on='id')
movies_df_merge.head()
```

```python
movies_df_merge.shape
```

```python
movies_cleaned_df=movies_df_merge.drop(columns=['homepage','title_x','title_y','status','production_countries'])
movies_cleaned_df.head()
```

```python
movies_cleaned_df.shape
```

```python
movies_cleaned_df.dtypes
```

```python
movies_cleaned_df.isnull().mean()
```

## Now we will use weighted average for each movie avg voting

## W = (Rv + Cm) / (v+m)

### W=Weighted rating
### R=avg rating of movie(bw 1 to 10)
### v=number of votes
### m=minimum votes reqd to be listed in top 200
### C=mean vote across whole report

```python
#create all above variables
```

```python
v=movies_cleaned_df['vote_count']
R=movies_cleaned_df['vote_average']
C=movies_cleaned_df['vote_average'].mean()
m=movies_cleaned_df['vote_count'].quantile(0.70)  #should have more than 70 percentile votes to be listed
```

## Based on vote avg and vote count, we have performed weighted avg

```python
movies_cleaned_df['weighted_avg']=((R*v)+(C*m))/(v+m) 
```

```python
movies_cleaned_df.head()
```

```python
movies_based_ranking=movies_cleaned_df.sort_values('weighted_avg',ascending=False)
```

```python
movies_based_ranking[['original_title','vote_count','vote_average','weighted_avg','popularity']].head()
```

## This is a basic recommendor system and it can be based on either collaborative filtering or content based filtering coz it is capturing user behaviour toward votes.

```python
sns.barplot(y=movies_based_ranking['original_title'].head(10),x=movies_based_ranking['weighted_avg'].head(10),data=movies_based_ranking)
```

## we have seen recommending using weighted avg but we should also consider popularity as many people dont see it but should be considered

```python
popularity=movies_based_ranking.sort_values('popularity',ascending=False)
```

```python
sns.barplot(x=popularity['popularity'].head(10),y=popularity['original_title'].head(10),data=popularity)
```

## We can see that according to weighted avg, shawshank redemption was the best movie but according to popularity, Minions is on top

## So we combine these two concepts of weighted avg and popularity to recommend movie

```python
#50% priority is given to each
```

```python
#first lets normalize both columns as the magnitude varies alot in both
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaling=MinMaxScaler()
```

```python
movies_scaled_df=scaling.fit_transform(movies_cleaned_df[['weighted_avg','popularity']])
```

```python
movie_normalized_df=pd.DataFrame(movies_scaled_df,columns=[['weighted_avg','popularity']])
```

```python
movie_normalized_df.head()
```

```python
movies_cleaned_df[['normalized_weighted_avg','normalized_popularity']]=movie_normalized_df
```

```python
movies_cleaned_df.head()
```

```python
movies_cleaned_df['score']=movies_cleaned_df['normalized_weighted_avg']*0.5 + movies_cleaned_df['normalized_popularity']*0.5
```

```python
movies_scores_df=movies_cleaned_df.sort_values(['score'],ascending=False)
```

```python
movies_scores_df[['original_title','normalized_weighted_avg','normalized_popularity','score']].head(20)
```

## so we have taken both weighted avg and popularity to get a final score

## So based on score, Interstellar will be given recommendation then Minions will be given recommendation and so on..

```python
sns.barplot(x=movies_scores_df['score'].head(10), y=movies_scores_df['original_title'].head(10), data=movies_scores_df)
```
